```python
import numpy as np
import pandas as pd
import plotly.express as px
import statsmodels.formula.api as smf
```

a), b), c)

```python
np.random.seed(1)
x = np.random.normal(size=100)
eps = np.random.normal(size=100, scale=0.25)
y = -1 + 0.5 * x + eps
data = pd.DataFrame(data={"x": x, "y": y})
```

- the length of y is 100.
- the value of $\beta_0$ should be -1
- the value of $\beta_1$ should be 0.5

d) and f)

```python
px.scatter(x=x, y=y, trendline="ols", trendline_color_override="red")
```

e)

```python
results= smf.ols("y ~ x", data=data).fit()
print(results.summary())
```

- $\hat{\beta}_0 = -0.96$
- $\hat{\beta}_1 = 0.52$

Both of these values are really close to $\beta_0$ and $\beta_1$ fonded before.

- As you can see in the chart, there is a trendline with a positive relationship

g)

```python
results_poly = smf.ols("y ~ x + I(x ** 2)", data=data).fit()
print(results_poly.summary())
```

The $R^2$ statistic was kept, the F-score decreased and p-value is higher, so the fit is worst than the previous one.
