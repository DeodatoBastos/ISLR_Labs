```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
path = "~/Desktop/ISLR_Labs/data/Auto.csv"
data = pd.read_csv(path)
```

```python
data.describe()
```

```python
data.info()
```

```python
data.sample(10)
```

Changing '?' mark to NaN to replace easily

```python
data.replace('?', np.nan, inplace=True)
data.dropna(inplace=True)
data["horsepower"] = pd.to_numeric(data["horsepower"])
data["origin"] = data["origin"].apply(str)
```

```python
data.info()
```

```python
quantitative = data.iloc[:, :7]

pd.DataFrame({'Min': quantitative.min(),
              'Max': quantitative.min(),
              'Mean': quantitative.mean(),
              'Std': quantitative.std()})
```

Remove the 10th through 85th observations

```python
filtered_data = data.drop(index=data.index[9:84])
```

```python
filtered_quantitative = filtered_data.iloc[:, :7]

pd.DataFrame({'Min': filtered_quantitative.min(),
              'Max': filtered_quantitative.min(),
              'Mean': filtered_quantitative.mean(),
              'Std': filtered_quantitative.std()})
```

```python
fig = px.scatter_matrix(data,
                        dimensions=data.columns[[i for i in range(7)]],
                        symbol="origin", color="origin",
                        title="Scatter matrix of Auto data set",
                        labels={col: col.replace('_', ' ') for col in data.columns},
                        height=960, width=1240)
fig.update_traces(diagonal_visible=False)
fig.show()
```

```python
data.corr(numeric_only=True)
```

The scatter plots showed us that has a linear relationship between some predictors.
For example, displacement and horsepower, displacement and weight, horsepower and weight. In others the relation is more complicated.

Furthermore, mpg has a strong negative correlation with displacement, horsepower and weight, what makes much sense.

Also, pay attention on the origin, that changes a lot the relationship between the predictors.
