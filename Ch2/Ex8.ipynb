```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "~/Desktop/ISLR_Labs/data/College.csv"
data = pd.read_csv(path)

data.rename(columns={"Unnamed: 0": "College"}, inplace=True)
```

```python
data.head()
```

```python
data.describe()
```

```python
axes = pd.plotting.scatter_matrix(data.iloc[:, :6], alpha=0.2)
plt.tight_layout()
```

```python
data.boxplot(column="Outstate", by="Private", grid=False)
```

```python
data["Elite"] = np.where(data["Top10perc"] > 50, "Yes", "No")
data.describe(include='all')
```

```python
data.boxplot(column="Outstate", by="Elite", grid=False)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].hist(data["Apps"], bins=25)
axs[0, 0].set_title('Applications')
axs[0, 0].set_xlim(0, 25000)

axs[0, 1].hist(data["Top10perc"])
axs[0, 1].set_title('New students from top 10 % of high school class')

axs[1, 0].hist(data["PhD"])
axs[1, 0].set_title('Percent of faculty with Ph.D.’s')

axs[1, 1].hist(data["Expend"])
axs[1, 1].set_title('Instructional expenditure per student')
```
