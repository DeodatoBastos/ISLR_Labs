```python
fig = px.scatter_matrix(data,
                        dimensions=["crim", "zn", "indus", "age", "dis",
                                    "tax", "medv"],
                        title="Scatter matrix of Boston data set",
                        labels={col: col.replace('_', ' ')
                                for col in data.columns},
                        height=960, width=1240)
fig.update_traces(diagonal_visible=False)
fig.show()
```
